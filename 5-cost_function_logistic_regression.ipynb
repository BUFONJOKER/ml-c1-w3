```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import plot_data, sigmoid, dlc
plt.style.use('./deeplearning.mplstyle')
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])


print(f"X_train = {X_train.shape}")
print(f"y_train = {y_train.shape}")
```

```python
fig, ax = plt.subplots(1, 1)
plot_data(X_train, y_train, ax)
ax.axis([0, 4, 0, 3.5])
ax.set_xlabel("$x_0$", fontsize=20)
ax.set_ylabel("$x_1$", fontsize=20)
plt.show()
```

```python
def compute_cost(X, y, w, b):
    
    m = X.shape[0]
    cost = 0.0
    
    for i in range(m):
        z_i = np.dot(X[i],w) + b
        fw_b_xi = sigmoid(z_i)
        cost+= -y[i] * np.log(fw_b_xi) - (1-y[i]) * np.log(1-fw_b_xi)
        
    cost /= m
    
    return cost
        
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
print(compute_cost(X_train, y_train, w_tmp, b_tmp))
```

```python
x0 = np.arange(0,6)
x1 = 3 - x0
x1_other = 4 - x0

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x0, x1, c=dlc['dlblue'], label='$b$=-3')
ax.plot(x0, x1_other, c=dlc['dlmagenta'], label='$b$=-4')
ax.axis([0, 4, 0, 4])

plot_data(X_train, y_train, ax)
ax.axis([0, 4, 0, 4])
ax.set_xlabel("$x_0$")
ax.set_ylabel("$x_1$")
plt.title("Decision Boundary")
plt.show()
```

```python
w_array_1 = np.array([1, 1])
b_1 = -3

w_array_2 = np.array([1, 1])
b_2 = -4

print(f"Cost for b = -3 : {compute_cost(X_train, y_train, w_array_1, b_1)} ")
print(f"Cost for b = -4 : {compute_cost(X_train, y_train, w_array_2, b_2)} ")
```
