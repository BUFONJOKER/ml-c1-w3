```python
def sigmoid(z):
    g = 1 / (1+np.exp(-z))
    return g
```

```python
z_tmp = np.arange(-10, 11)

y = sigmoid(z_tmp)

print("Input z  Output y")
np.set_printoptions(precision=3)
print(np.c_[z_tmp, y])
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z_tmp,y, c='b')
ax.set_title("Sigmoid Function")
ax.set_xlabel("z")
ax.set_ylabel("sigmoid(z)")
draw_vthresh(ax,0)
plt.show()
```
