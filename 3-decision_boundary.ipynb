```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import sigmoid,draw_vthresh, plot_data
plt.style.use('./deeplearning.mplstyle')
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

```python
print(np.array([0,1,2,3,4]).shape)
print(np.array([0,1,2,3,4]).reshape(-1,1).shape)
print(X.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4,4))
ax.axis([0, 4, 0, 3.5])
plot_data(X, y, ax)
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

```python
z = np.arange(-10, 11)
fig, ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z, sigmoid(z), c='b')
ax.set_title('Sigmoid Function')
ax.set_xlabel('z')
ax.set_ylabel('sigmoid(z)')
draw_vthresh(ax, 0)
```

```python
x0 = np.arange(0,6)
x1= 3 - x0
fig, ax = plt.subplots(1,1, figsize=(5,3))
ax.plot(x0, x1, c='b')
ax.axis([0,4,0,3.5])
ax.fill_between(x0, x1, alpha=0.2)
plot_data(X, y, ax)
ax.set_xlabel(r'$x_0$')
ax.set_ylabel(r'$x_1$')
plt.show()
```
