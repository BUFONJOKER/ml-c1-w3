```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from lab_utils_common import dlc, plot_data
from plt_one_addpt_onclick import plt_one_addpt_onclick
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0, 0, 0, 1, 1, 1])
x_train_2 = np.array([
    [0.5, 1.5],
    [1, 1],
    [1.5, 0.5],
    [3, 0.5],
    [2, 2],
    [1, 2.5]
])
y_train_2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
positive = y_train == 1
negative = y_train == 0

fig, ax = plt.subplots(1, 2, figsize=(8,3))
ax[0].scatter(x_train[positive], y_train[positive],
               marker='x', s = 80, c='red', label='y=1')

ax[0].scatter(x_train[negative], y_train[negative],
               marker='o', s = 80, c='blue', label='y=0')

ax[0].set_ylim(-0.08, 1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one variable plot')
ax[0].legend()

plot_data(x_train_2, y_train_2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_ylabel('$x_1$',fontsize=12)
ax[1].set_xlabel('$x_0$',fontsize=12)
ax[1].set_title('two variable plot')
ax[1].legend()
plt.tight_layout()
plt.show()
```
